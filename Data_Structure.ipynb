#Data Types and Structures Questions

1. What are data structures, and why are they important?
  - Data structures organize, manage, and store data efficiently, making processing and retrieval faster.


2. Explain the difference between mutable and immutable data types with examples.
  - Mutable types of data can be changed whereas immutable types cannot be changed after creation.


3. What are the main differences between lists and tuples in Python?
  - Lists are mutable, slower, and use more memory whereas tuples are immutable, faster, and memory-efficient.


4. Describe how dictionaries store data.
  - Dictionaries store data as key-value pairs using a hash table for fast lookup.


5. Why might you use a set instead of a list in Python?
  - Sets automatically remove duplicates and provide faster membership testing.


6. What is a string in Python, and how is it different from a list?
  - A string is an immutable sequence of characters; a list is mutable and can hold any data type.


7. How do tuples ensure data integrity in Python?
  - Tuples are immutable, preventing accidental changes to the data.


8. What is a hash table, and how does it relate to dictionaries in Python?
  - A hash table maps keys to values using hashing; dictionaries use this structure internally.


9. Can lists contain different data types in Python?
  - Yes, lists can contain elements of any and mixed data types.


10. Explain why strings are immutable in Python.
  - Immutability allows memory optimization and protects data from unintended changes.


11. What advantages do dictionaries offer over lists for certain tasks?
  - Faster lookup using keys, better for associative data, and easier to read.


12. Describe a scenario where using a tuple would be preferable over a list.-When storing fixed values like coordinates or days of the week that should not change.


13. How do sets handle duplicate values in Python?
  - Sets discard duplicates and store only unique elements.


14. How does the “in” keyword work differently for lists and dictionaries?
  - In lists, it checks for values; in dictionaries, it checks for keys.


15. Can you modify the elements of a tuple? Explain why or why not.
  - No, because tuples are immutable, meaning their content cannot be changed after creation.


16. What is a nested dictionary, and give an example of its use case?
  - A dictionary inside another dictionary. Example: login = {'credentials': {'name': 'yash', 'password': 2543}}.


17. Describe the time complexity of accessing elements in a dictionary.
  - O(1) on average due to hash-based key access.


18. In what situations are lists preferred over dictionaries?
  - When order is important or when only sequential data is needed , we prefer list.


19. Why are dictionaries considered unordered, and how does that affect data retrieval?
  - Dictionaries did not preserve order. If they do, keys must be known for access.


20. Explain the difference between a list and a dictionary in terms of data retrieval.
  - Lists use index positions; dictionaries use unique keys for retrieving values.
